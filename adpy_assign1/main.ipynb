```python
# Math module provides various mathematical and trigonometric functions
import math
print(math.sqrt(361))
print(math.sin(30)) #radian
print(math.pi) # value of pi
theta = 30*math.pi/180 # Convert 30-degree to radian
print(math.sin(theta)) # sine of 30-degree
b = math.asin(0.5) # radian
# Convert the value of b from radian to degree
theta = b*180/math.pi
print(theta)
# Alternatively we can convert to radians using math function
theta = math.radians(30)
print(math.sin(theta)) # It will produce the same result as previous one.
# In the python cell type math followd by dot (math.) and hit **Tab** It will show all the possible functions
```

```python
# Create an empty list
phonebook = []
# create an empty dictionary
phone = {}
# Create a loop for two times
for i in range(2):
    name = input('Write your name: ')
    num = input('What is your number?: ')
    
    # create a dictionary element
    phone['name'] = name
    phone['number'] = num
    # Add phone to the phonebook
    phonebook.append(phone)
print(phonebook)


```

```python
# Create an empty list
studentlist = []
# Create a dictionary of students
student = {'name': 'Ram Thapa', 'roll': 20, 'course' : 'Civil Engineering'}
studentlist.append(student) # Add the student to the list
# New student
student = {'name': 'Hari Pande', 'roll': 25, 'course' : 'Mechanical Engineering'}
studentlist.append(student) # Add again
print(studentlist)
```

```python
# Now let us loop through the list
for st in studentlist: # Each item of the list will be place in *st* variable
    print(f"Name: {st['name']}") # anything inside curly bracket will be evaluated and printed
    # We need to add *f* before the intended output
    print(f"Roll Number: {st['roll']:03d}") # The roll number is printed 
    print(f"Course: {st['course']}")
```
